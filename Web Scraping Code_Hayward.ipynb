```python
import urllib.request 
import urllib.parse
import urllib.error
import ssl
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as BS

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
url = "https://www.mlslistings.com/Search/Result/9fbbd96d-0150-4653-b8f2-8844f745ab31/1"
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()

soup = BS(webpage,'html.parser')
print(soup.title)
```

```python
add = soup.findAll('small',attrs={'class':'font-weight-bold'})
#print(int(add[0].get_text().split()[-2])
```

```python

total = int(add[0].get_text().split()[-2])
print(total)
links = []
found = 0;
i = 1
while (found < total):
    print(i)
    url = "https://www.mlslistings.com/Search/Result/9fbbd96d-0150-4653-b8f2-8844f745ab31/" + str(i)
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    soup = BS(webpage,'html.parser')
    listings = soup.find_all('h5',attrs={'class':'card-title font-weight-bold listing-address mb-25'})
    for list in listings:
        links.append(list.a['href'])
    print(len(listings))
    found = found + len(listings)
    i = i + 1
```
